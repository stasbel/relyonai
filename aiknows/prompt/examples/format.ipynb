```python
from aiknows import gpt
from aiknows.runtime import finish_task_ok, finish_task_error

from aiknows.runtime import setup_task
x = 10
setup_task(
    task='calculate x^x',
    args={'x': x},
    globals=globals(),
    validator=lambda r: r == 10000000000,
)
del setup_task
```

```python
finish_task_ok(x ** x);
```
